## HW1_2(a)
beacause $\frac{b[0]}{1 + a[1]Z^-1} = \frac{Y[n]}{X[n]}$   so y[n] = -a[1]y[n-1] + b[0]x[n]
<br>
And unit gain at DC means that $\frac{b[0]}{1 + a[1]} = 1$   then b[0] = 1 + a[1]
<br>
The pole $\alpha$is a value of z where H(z) goes to infinity . so $1 + a[1]Z^-1$ = 0  so $\alpha$ = -a[1] and b[0] = 1 - $\alpha$
<br>
In all, **a[1]** = -$\alpha$ and **b[0]** = 1 - $\alpha$

## HW1_2(b)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
alpha = [0.9,0.5,0.1,-0.5]
aa = [-a for a in alpha]
bb =[1-a for a in alpha]
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2,figsize=(12,12))
ax = [ax1,ax2,ax3,ax4]
for i in range(4):
    w, h =signal.freqz(b = bb[i],a = [1,aa[i]])
    ax[i].plot(w, 20 * np.log10(abs(h)), 'b')
    ax[i].set_xlim(0,2*np.pi*0.5)
    ax[i].set_ylabel('Amplitude [dB]', color='b')
    ax[i].set_xlabel('Frequency [rad/sample]')
    ax[i].set_title(('alpha = {}').format(alpha[i]))
```

## HW1_2(c)
### From the three graphs , we can know that for when $\alpha$ is 0.9,0.5,0.1, the time constant is 16, 3, 1

```python
alpha = [0.9,0.5,0.1]
fig,(ax1,ax2,ax3) = plt.subplots(1,3,figsize=(12,6))
ax = [ax1,ax2,ax3]
# From the graph, we can see the time constant, so I clarify here 
##to make it more convincing 
xx = [16, 2.25, 0.73]
for i in range(3):
    num = 1 - alpha[i]
    den = [1,-np.log(alpha[i])]
    sys = signal.TransferFunction(num, den)
    t , y = signal.impulse(system = sys)
    ax[i].plot(t,y)
 #   ax[i].plot([ ( alpha[i] ** n ) * (1 - alpha[i] )  for n in range(20) ] )
    ax[i].axhline(y = 0.2*y[0],color='g', linestyle='--')
    # Add the time constant verticle line to make it more clear
    ax[i].axvline(x = xx[i] ,color='g', linestyle='--')
    ax[i].annotate(xx[i],(xx[i],-0.04*y[0]),color = 'r')
    ax[i].set_ylabel('impulse response h[n]')
    ax[i].set_xlabel('n')
    ax[i].set_title(('alpha = {}').format(alpha[i]))
```
